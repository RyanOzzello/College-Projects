```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import string
import cv2
```

```python
# read in an output file from tesseract
def read_tess(file):
    file_ob = open(file,"r",encoding="utf8")
    tess = file_ob.readlines()
    for i in range(0,len(tess)):
        tess[i] = re.sub('','',tess[i]).strip('\n')
    return tess

#check is value is a number
#int
def is_number_int(s):
    try:
        int(s)
        return True
    except ValueError:
        return False
#or float
def is_number_float(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
    
#filter out anything that is not a number
#int
def filter_digits_int(labeled):
    filtered = []
    for r in range(0,len(labeled)):
        if is_number_int(labeled[r][1]):
            filtered.append((labeled[r][0],int(labeled[r][1])))
    return filtered
#float
def filter_digits_float(labeled):
    filtered = []
    for r in range(0,len(labeled)):
        if is_number_float(labeled[r][1]):
            filtered.append((labeled[r][0],float(labeled[r][1])))
    return filtered


#remove outliers in data that are read incorrectly
def remove_outliers(data,threshold):
    clean = data.copy()
    i = 0
    while i < len(clean)-1:
        if abs(int(clean[i][1]) - int(clean[i+1][1])) > threshold:
            del clean[i+1]
        else:
            i += 1
    return clean

#convert the frames to seconds into the video
def frames_to_seconds(data):
    for i in range(0,len(data)):
        data[i] = (data[i][0] / 5, data[i][1])
    return data
```

```python
#speed 

speedFile = "data/outputSpeed.txt"

#read in tesseract output on speed data
speedRaw = read_tess(speedFile)

#keep track of frames
speedFrames = np.array(range(0,len(speedRaw)))
#join frames to data
speedLabeled = list(zip(speedFrames,speedRaw))

#filter out non numbers
speedDigits = filter_digits_int(speedLabeled)

#remove outliers and tess readings to ints
speedFiltered = remove_outliers(speedDigits,500)

#change frames to seconds
speedTime = frames_to_seconds(speedFiltered)

#break into x and y to plot frames in seconds
speedTimeDf = pd.DataFrame(speedTime, columns=["second","speed"])
speedXs = np.array(speedTimeDf['second'])
speedYs = np.array(speedTimeDf['speed'])

#plot speed over time
print("Total Frames extracted: " + str(len(speedRaw)))
print("Total usable frames: " +  str(len(speedFiltered)))
print("Frames Lost: " + str((len(speedRaw)-len(speedFiltered))))

plt.plot(speedXs,speedYs);
plt.ylabel("km/h");
plt.xlabel("seconds into launch");
```

```python
# make dictionary of names of teams for easier identification in labeling game number and team names
teamDict = {'Raptors':1, 'Hornets':2,'Nets':3,'Lakers':4,'76ers':5,'Hawks':6,'Wolves':7, 'Bulls':8}

#one giant clean function for the scores
def clean_scores(labeled):
    clean = []
    
    #use table to remove punctuation from scores
    table = str.maketrans('', '', string.punctuation)
    
    #go through your data of scores and their frames
    for i in range(0,len(labeled)):
        #remove punctuation in score read
        noPunc = labeled[i][1].translate(table)
        #split into words by spaces
        cleaned = noPunc.split(" ")
        #if array has 4 components then it is in the proper form("team, points, team, points")
        if len(cleaned) == 4:   
            #if the 2nd and 4th components are numbers then we also know it is in a good form
            if is_number_int(cleaned[1]) and is_number_int(cleaned[3]):
                #add to clean list
                clean.append((labeled[i][0],noPunc))
    
    #do further cleaning:
    #this section looks at all of the cleaner scores for each games and choses the score reading 
    #that showed up the most for each game 
    #ex: 1)"Raptors 100 Hornets 56", 2)"Raptors 100 Hornets 56", 3)"Rapto 10 Hornets 5":
    # this code would see that the reading in 1 and 2 is the most common reading and 
    #chooses this reading as the final cleaned score reading
    cleanScores = []
    for i in range(0,len(clean)):
        cleanScores.append(clean[i][1])
    #set of each unique reading
    setScores = set()
    for i in cleanScores:
        setScores.add(i)
    setScores = list(setScores)
    counts = []
    for i in setScores:
        counts.append((cleanScores.count(i),i))
    #create a dataframe with the score reading and the frequency of that reading sorted by frequency desc
    countsDf = pd.DataFrame(counts,columns=['count','score'])
    countsDf = countsDf.sort_values('count',ascending=False)
    gameNum = []
    #label the readings by game number so we know which readings are for which game
    # and go from the highest frequency score reading to the lowest storing one 
    #reading from each game of the score reading that had the greatest frequency
    for i in countsDf['score']:
        for name in teamDict:
            if name[0:3] in i[0:3]:
                gameNum.append(teamDict[name])
    countsDf['gameNum'] = gameNum
    chosenScores = []
    gamesVisited = []
    for i in range(0,len(countsDf)):
        if countsDf['gameNum'][i] not in gamesVisited:
            chosenScores.append(countsDf['score'][i])
            gamesVisited.append(countsDf['gameNum'][i])

    return chosenScores
        
# get the winners by comparing the scores
def winners(scores):
    winnerList = []
    for i in range(0,len(scores)):
        parts = scores[i].split(" ")
        if int(parts[1]) > int(parts[3]):
            winnerList.append((scores[i], parts[0]))
        else:
            winnerList.append((scores[i],parts[2]))
    return winnerList
```

```python
#Sports Ball scores
scoreFile = "data/outputSBScore.txt"

#read in tesseract output on score data
scoreRaw = read_tess(scoreFile)
#keep track of frames
scoreFrames = np.array(range(0,len(scoreRaw)))
#join frames to data
scoreLabeled = list(zip(scoreFrames,scoreRaw))
#clean
cleanestScores = clean_scores(scoreLabeled)
#get winners
scoresWinners = winners(cleanestScores)

print("Readings Correct: 4/8")
print("Winners Correct: 5/8")
print()
for i in scoresWinners:
    print("Game:" + i[0] + "      Winner:" + i[1])
print()
print("Game scores read correctly:")
print("Rapters: Incorrect (Pistons 110)")
print("Rapters: Incorrect (Pistons 110)")
print("Nets:    Correct")
print("Hawks:   Incorrect (Magic 101)")
print("Bulls:   Correct")
print("Hornets: Correct")
print("76ers:   Incorrect (Bucks 125)")
print("Wolves:  Incorrect (Rockets 117)")
print("Lakers:  Correct")
```

```python
# create dictionary for clean_titles
titleDict = np.array(["MIDWEST","WEST","TENNIS","SOUTH","MLB","MLS","NFL","NCAAW","NCAAM","GOLF","NHL","NBA"])
#if exact title in dictionary(array) is found in the reading, chose that reading as a good one
def clean_titles(labeled):
    goodTitles = []
    for i in range(0,len(labeled)):
        for t in titleDict:
            if t in labeled[i][1]:
                goodTitles.append((labeled[i][0],t))
                break
    return goodTitles

#cleaning attempts for another part of the banner
def clean_text(labeled):
    #table for removing punctuation
    table = str.maketrans('', '', string.punctuation)
    for i in range(0,len(labeled)):
        cleaned = labeled[i][1].translate(table)
        #split in array, remove empty strings, and remove single length strings
        cleaned = cleaned.split(" ")
        l = 0
        while l < len(cleaned):
            cleaned[l] = cleaned[l].strip()
            if len(cleaned[l]) == 1:
                del cleaned[l]
            else:
                l += 1
        #rejoin into one single string
        cleaned = " ".join(cleaned)
        labeled[i] = (labeled[i][0],cleaned)
    j = 0    
    #remove shorter texts since unintelligable 
    while j < len(labeled):
        if labeled[j][1] == '' or len(labeled[j][1]) < 6:
            del labeled[j]
        else:
            j += 1
    return labeled
```

```python
#Sport Ball Title,Info,Text of bottom banner
#Title: label of the banner (WEST,TENNIS,MIDWEST,SOUTH,MLB,MLS,NFL FA,NCAAW,NCAAM,GOLF,NHL,NBA)
#Info: scores, teams, what game
#Text: what division, fun stats

sbTitleFile = "data/outputSBTitle.txt"
sbInfoFile = "data/outputSBInfo.txt"
sbTextFile = "data/outputSBText.txt"

#titles
#read in tesseract output
titleRaw = read_tess(sbTitleFile)
#keep track of frames
titleFrames = np.array(range(0,len(titleRaw)))
#join frames to data
titleLabeled = list(zip(titleFrames,titleRaw))
#clean titles
titleClean = clean_titles(titleLabeled)

#info
#read in tesseract output
infoRaw = read_tess(sbInfoFile)
#keep track of frames
infoFrames = np.array(range(0,len(infoRaw)))
#join frames to data
infoLabeled = list(zip(infoFrames,infoRaw))
#clean info
infoClean = clean_text(infoLabeled)

#text
#read in tesseract output
textRaw = read_tess(sbTextFile)
#keep track of frames
textFrames = np.array(range(0,len(textRaw)))
#join frames to data
textLabeled = list(zip(textFrames,textRaw))
#clean info
textClean = clean_text(textLabeled)

#put the banner together with accessible info
textCleanDf = pd.DataFrame(textClean, columns=['frames','text'])
titleCleanDf = pd.DataFrame(titleClean, columns=['frames','title'])
infoLabeledDf = pd.DataFrame(infoLabeled, columns=['frames','info'])
organized = titleCleanDf.merge(infoLabeledDf)
organized = organized.merge(textCleanDf)
print("Total Frames extracted: " + str(len(textRaw)))
print("Total usable frames: " +  str(len(textClean)))
print("Frames Lost: " + str((len(textRaw)-len(textClean))))
organized
```
