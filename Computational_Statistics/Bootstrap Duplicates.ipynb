# Duplicate Percentage of Bootstrapping as N Increases

## Task
We want to explore the method of bootstrapping by evaluating the percentage of duplicates. We can see that the number of duplicates converges to about 1/3 as N increases.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 200
avgs = np.array([])

for n in range(2, N):
    inc_avg_dup = 0
    D = np.array(range(n))
    for i in range(1000):
        d = np.random.choice(D, n, replace=True)
        num_dups = len(d) - len(set(d))
        inc_avg_dup = inc_avg_dup + (1/(i+1)) * (num_dups - inc_avg_dup)
        
    avgs = np.append(avgs, inc_avg_dup)
```

```python
plt.title('Number of duplicates as N increases')
plt.xlabel('N')
plt.ylabel('Total Duplicates')
plt.plot(avgs);
```

```python
data_size = np.array([range(2,N)])
avgs_perc = (avgs / data_size)[0]
avgs_perc = np.append(np.array([0,0]), avgs_perc)

plt.title('Percentage of duplicates as N increases')
plt.xlabel('N')
plt.ylabel('Percentage Duplicates')
plt.plot(avgs_perc);
```
